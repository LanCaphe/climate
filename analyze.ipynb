```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df= pd.read_csv('final+filters.csv',low_memory=False)
```

```python
#display the rows where city is missing
df[df['city'].isnull()]
```

```python
# display all differents regions
df['cdp_region'].unique()
```

```python
# change the values of cdp_region if it is South Asia or South East Asia
df['cdp_region'] = df['cdp_region'].replace(['South Asia', 'South East Asia', 'East Asia', 'South Asia','South and West Asia', 'Southeast Asia'], 'Asia')
df['cdp_region'] = df['cdp_region'].replace('Middle east', 'Middle East')
df['cdp_region'] = df['cdp_region'].replace(['Canada', 'United States of America'],'North America')
df['cdp_region'] = df['cdp_region'].replace('South America','Latin America')
```

```python
df[df['cdp_region'] == 'Southeast Asia and Oceania']
```

```python
asia_countries = ['Malaysia', 'Philippines', 'Thailand','Viet Nam', 'Indonesia' ]

# Change the values of cdp_region for Asian countries
df.loc[df['country'].isin(asia_countries), 'cdp_region'] = df['cdp_region'].replace('Southeast Asia and Oceania', 'Asia')

# Change the value of cdp_region for Australia
df.loc[df['country'] == 'Australia', 'cdp_region'] = df['cdp_region'].replace('Southeast Asia and Oceania', 'Oceania')
```

```python
df['cdp_region'].unique()
```

```python
columns_graph_1 = ['reporting_year', 'cdp_region', 'city', 'organization_number', 'emissions_reported'] 
df_graph_1 = df[columns_graph_1]

#remove duplicates
df_graph_1 = df_graph_1.drop_duplicates(subset=['organization_number', 'reporting_year'], keep="first")
df_graph_1
```

```python
df_graph_1['cdp_region'].unique()
```

```python
# display the rows with cdp_region = 'Southeast Asia and Oceania'
df_graph_1[df_graph_1['cdp_region'] == 'Southeast Asia and Oceania']
```

```python
grouped = df_graph_1.groupby(['reporting_year','cdp_region']).sum()
grouped 

```

```python
# remove columns city and organization_number
grouped = grouped.drop(['organization_number'], axis=1)
grouped
```

```python
#create a dataframe with the total emissions reported per year
total_emissions = grouped.groupby(['reporting_year']).sum()

#create a color palette persolnalized for seaborn barplot
colors = ["#1a233a", "#009887","#f7e33f","#004776","#8fbeb5","#cbdfdb"]

#create a barplot with the total emissions reported per year and per region with seaborn with the personalized color palette , the title is the total emissions reported per year
ax = sns.barplot(x="reporting_year", y="emissions_reported", hue="cdp_region", data=grouped.reset_index(), palette=colors)

#set the title of the graph
ax.set_title('Total emissions reported per year and per region')
```
